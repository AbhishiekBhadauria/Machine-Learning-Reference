## Handling Missing Values

```python
data.info()
```

```python
pd.isnull(data['Country']).sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
mean_imputer = SimpleImputer(strategy='mean')
mode_imputer = SimpleImputer(strategy='most_frequent')
```

```python
data
```

```python
data['Country'] = mode_imputer.fit_transform(data['Country'].values.reshape(-1, 1))
data['Age'] = mean_imputer.fit_transform(data['Age'].values.reshape(-1, 1))
data['Salary'] = mean_imputer.fit_transform(data['Salary'].values.reshape(-1, 1))
```
